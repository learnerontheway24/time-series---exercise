# （d）：检验零假设：每个股票的对数收益率均值为0

```python
import scipy.stats as ss
```

```python
t,p_twotail=ss.ttest_1samp(stock_log, 0)  #双尾假设检验，均值t检验，前面的是需要被假设检验的数据，后面是假设检验均值是否等于的那个值
```

```python
t,p_twotail   #上面一行是t值（统计量的值），下面一行是对应的p值
```
